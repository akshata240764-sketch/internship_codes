```python
#Question 2: Sum of Even Numbers
#Write Python code that iterates through the following list numbers = [12, 7, 23, 48, 15, 66, 9]. Calculate and print the sum of all even numbers in the
#list.
numbers=[12,7,23,48,15,66,9]
sum=0
for i in numbers:
    if i%2==0:
        sum+=i
print(sum)
```

```python
#Question 3: Manual List Reversal
#Given a list data = ['apple', 'banana', 'cherry', 'date', 'elderberry'], create a new list that contains the elements of data in reverse order. You are
#not allowed to use data.reverse() or data[::-1]. You must achieve this using loops and basic list operations.
data=['apple','banana','cherry','date','elderberry']
reversed_list=[]
for items in data:
    reversed_list.insert(0,items)
print(reversed_list)
```

```python
#Question 4: Finding the Maximum Element
#Without using the built-in max() function, write code to find and print the largest number in the list values = [8, 2, 10, 5, 17, 3, 14].
values=[8,2,10,5,17,3,14]
largest=values[0]
for i in values:
    if i>largest:
        largest=i
print(largest)
    
```

```python
#Question 5: Count Occurrences
#Given the list items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red'], write code to count how many times the string 'red' appears in the list
#and print the count. Do not use the list.count() method.
items=['red','blue','green','red','yellow','blue','red']
count=0
for item in items:
    if item=='red':
        count+=1
print('count of red:',count)
    
```

```python
# Question 6: Remove All Occurrences of an Element
# Given a list mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5], write code to remove all instances of the number 5 from the list. The final list should only
# contain [1, 2, 3, 4]. You must achieve this by iterating and creating a new list or modifying the existing one carefully.
mixed_numbers1=[1,5,2,5,3,5,4,5]
for i in mixed_numbers1:
    if i==5:
        mixed_numbers1.remove(i)
print(mixed_numbers1)
    
```

```python
# Question 7: List Palindrome Check
# Write code to determine if a given list is a palindrome (reads the same forwards and backwards). Print "Palindrome" if it is, and "Not a Palindrome" otherwise.
# Test with:
# ● list1 = [1, 2, 3, 2, 1]
# ● list2 = ['a', 'b', 'c', 'b', 'd'] You are not allowed to use [::-1] for comparison.
list1=[1,2,3,2,1]
list2=['a','b','c','b','d']
for data in [list1,list2]:
    reversed_list=[]
    for item in data:
        reversed_list.insert(0,item)
    if data==reversed_list:
        print("palindrome")
    else:
        print("not a palindrome")
```

```python
# Question 8: Merge and Sort Two Lists (Manual Sort)
# Given two lists:
# ● list_a = [3, 1, 4]
# ● list_b = [2, 5, 0] Combine these two lists into a single list and then sort the combined list in ascending order without using the list.sort() method
# or sorted() function. You'll need to implement a basic sorting algorithm (e.g., bubble sort logic) using loops
list_a=[3,1,4]
list_b=[2,5,0]
combined=list_a+list_b
for i in range(len(combined)):
    for j in range(len(combined)-1):
        if combined[j]>combined[j+1]:
            combined[j],combined[j+1]=combined[j+1],combined[j]
print(combined)
```

```python
# Question 9: Flatten a Nested List
# You are given a list that contains integers and other lists (only one level of nesting). For example: nested_list = [1, [2, 3], 4, [5, 6, 7], 8].
# Write code to create a new flat list containing all the integers from nested_list. The result should be [1, 2, 3, 4, 5, 6, 7, 8].
nested_list=[1,[2,3],4,[5,6,7],8]
sorted=[]
for item in nested_list:
    if type(item)==list:
        for num in item:
            sorted.append(num)
    else:
        sorted.append(item)
print(sorted)
```

```python
# Question 10: Find Common Elements
# Given two lists, list_p = [1, 2, 3, 4, 5] and list_q = [4, 5, 6, 7, 8], write code to find and print all elements that are common to both lists. The
# order of common elements does not matter. You are not allowed to use set operations.
list_p=[1,2,3,4,5]
list_q=[4,5,6,7,8]
common=[]
for i in list_p:
    for j in list_q:
        if i==j:
            common.append(i)
print(common)
```
