```python
#Accept a temperature in Celsius from the user, convert it to Fahrenheit and Kelvin, and
#print all three values with labels.
celsius=int(input("enter the temperature in celsius:"))
print("celsius:",celsius)
print("fahrenheit:",(9/5*celsius)+32)
print("kelvin:",273.5+celsius)
```

```python
#Build a simple calculator — take two numbers and an operator (+, -, *, /) as input,
#perform the operation, and print the result. Print an error message for invalid operators or
#division by zero.
num1=int(input("enter first no:"))
num2=int(input("enter second no:"))
operator=input("enter the operator:")
if operator == "+":
    print(num1+num2)
elif operator == "-":
    print(num1-num2)
elif operator == "*":
    print(num1*num2)
elif operator == "/":
    if num2 == 0:
        print("cannot divide")
    else:
        print(num1/num2)
else:
    print("invalid operator")
```

```python
#8. Take a positive integer n as input and print all prime numbers from 2 to n.
n=int(input("enter a number:"))
for num in range(2,n+1):
    is_prime=True
    for i in range(2,num):
        if num%i==0:
            is_prime=False
            break
    else:
        print(num)
```

```python
#Keep accepting numbers from the user until they enter 0. Then print the total count of
#numbers entered, their sum, and their average. Do not include 0 in the calculations
count=0
total=0
while True:
    num=int(input("enter a number:"))
    if num==0:
        break
    count+=1
    total+=num
print("count:",count)
print("sum:",total)
print("average:",total/count)
```

```python
#Take n as input and print a triangle pattern of n rows made of asterisks, where row i
#has (2i - 1) stars centered appropriately.
n=int(input("enter the number"))
for i in range(1,n+1):
    spaces=n-i
    stars=2*i-1
    print(" "*spaces+"*"*stars)
    
```
