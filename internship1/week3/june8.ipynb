```python
class Person:
    def __init__ (self,name,age,ht):
        self.name=name
        self.age=age
        self.height=ht
    def show_info(self):
        return f"name->{self.name},age->{self.age},height->{self.height}"
```

```python
p1=Person("jatin",20,5.8)
p2=Person("sachin",22,5.9)
```

```python
class Person:
    def __init__ (self,name,age,ht):
        self.name=name
        self.__age=age
        self.height=ht
    def show_info(self):
        return f"name->{self.name},age->{self.__age},height->{self.height}"
```

```python
p1=Person("jatin",20,5.8)
p2=Person("sachin",22,5.9)
```

```python
p1.show_info()
```
