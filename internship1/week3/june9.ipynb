```python
class Person:
    insti="MITU"
    def __init__ (self,age,ht):
        self.age=age
        self.ht=ht
    def show(self):
        print(f"age->{self.age},height->{self.ht},institute->{self.insti}")
```

```python
class menuapp:
    def __init__ (self):
        self.itemlist=[]
    def show_menu(self):
        print("menu\n1.add items\n2.view items\n3.exit")
    def add_item(self):
        item=input("enter the item name")
        self.itemlist.append(item)
    def view_items(self):
        return self.itemlist
    def run(self):
        while True:
            self.show_menu()
            ch=int(input("enter the choice"))
            if ch==1:
                self.add_item()
            elif ch==2:
                print(self.view_items())
            elif ch==3:
                break
            else:
                return "wrong entry"
```

```python
atharv=menuapp()
```

```python
atharv.run()
```

```python
class bank:
    def __init__(self):
        self.balance=0
    def check_balance(self):
        print("current balance",self.balance)
    def deposit_money(self):
        amount=float(input("enter amount to deposit"))
        self.balance+=amount
        print("amount deposited successfully")
    def withdraw_money(self):
        amount=float(input("enter amount to withdraw"))
        if amount<=self.balance:
            self.balance-=amount
            print("amount has been withdrawn")
        else:
            print("insufficient balance")
    def run(self):
        while True:
            print("1.check balance")
            print("2.deposit money")
            print("3.withdraw money")
            print("4.exit")
            choice=int(input("enter your choice"))
            if choice==1:
                self.check_balance()
            elif choice==2:
                self.deposit_money()
            elif choice==3:
                self.withdraw_money()
            elif choice==4:
                break
            else:
                print("wrong entry")
op=bank()
op.run()
```
